```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
```

```python
print(torch.device("cpu"))  # Ensure it's using CPU
```

```python
# Load model in CPU mode
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipe.to("cpu")  # Force CPU execution
```

```python
# Generate an image
prompt = "a futuristic wooden smart home with embedded IoT"
image = pipe(prompt).images[0]
# image.show()  # Show image
```

```python
# Convert image to NumPy array
image_np = image.convert("RGB")  # Ensure it's in RGB format
```

```python
# Display with Matplotlib
plt.imshow(image_np)
plt.axis("off")  # Hide axes
plt.show()
```
